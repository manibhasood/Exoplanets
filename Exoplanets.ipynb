```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Visualizes all the columns
pd.set_option('display.max_columns', None)
```

```python
#Import dataset
df = pd.read_csv('/content/exoplanet.csv')

#Print the shape of the dataset
print(df.shape)

#Select the top of the dataset
df.head()
```

```python
#Analysing the type of data
df.info()
```

```python
#Checking for NULL values
df.isna().any()
```

```python
#Replaceing NULL values from dataset with Nan
missing_values=['??','na','X','999999']
df=df.replace(missing_values,np.NaN)
```

```python
#Checking shape post removal of NULL values
df.shape
```

```python
#Function to clean dataset
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```
